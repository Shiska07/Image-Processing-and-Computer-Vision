```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
crunch = cv2.imread('crunch.jpg',0)  
```

```python
display(crunch)
```

```python
cereals = cv2.imread('cereal-aisle.png',0)
display(cereals)
```

```python
orb = cv2.ORB_create()
```

```python
kp1,des1 = orb.detectAndCompute(crunch,None)
kp2,des2 = orb.detectAndCompute(cereals,None)
```

```python
#creating a brute force matching object
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)
```

```python
matches = bf.match(des1,des2)
```

```python
#distance shows how good of a match it was 
matches[0].distance
```

```python
matches = sorted(matches, key = lambda x: x.distance)
```

```python
crunch_matches = cv2.drawMatches(crunch,kp1,cereals,kp2,matches[:26],None, flags = 2)
```

```python
display(crunch_matches)
# we can see that the orb brute force method did not do so well
```

```python
# sift descriptor: works well for images with differing size

sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1,des1 = sift.detectAndCompute(crunch,None)
kp2,des2 = sift.detectAndCompute(cereals,None)
```

```python
# we are still using brute force matcher but note now kp1 and kp2 were calculated using sift
bf = cv2.BFMatcher()
```

```python
matches = bf.knnMatch(des1,des2,k=2)      #finding the top 2 matches
```

```python
matches  # <first best match>    <second best match>   and each row is a feature
```

```python
'''
if the first and second best match are farther away from each other for a feature, the feature is not considered good 
for measuring matches. Therefore, a ratio test can be done to filter good features.
LESS DISTANCE == BETTER MATCH
'''

good_matches = []

for match1,match2 in matches:
    if match1.distance < 0.75*match2.distance:   #match1 better than match2 by 25% or less distance
        good_matches.append([match1])
```

```python
# we are left with the best matches
print(len(matches))
print(len(good_matches))
```

```python
sift_matches = cv2.drawMatchesKnn(crunch,kp1,cereals,kp2,good_matches,None,flags = 2)
```

```python
display(sift_matches)

#this was a way better match
```

```python
# FLANN(Fast Library for Approximating Nearest Neighbors) based matcher instead of brute force

sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1,des1 = sift.detectAndCompute(crunch,None)
kp2,des2 = sift.detectAndCompute(cereals,None)
```

```python
# defining flann parameters
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks = 50)
```

```python
flann = cv2.FlannBasedMatcher(index_params,search_params)
```

```python
matches = flann.knnMatch(des1,des2,k = 2)
```

```python
matches_mask = [[0,0] for i in range(len(matches))]
```

```python
matches_mask
```

```python

for i,(match1,match2) in enumerate(matches):
    
    if match1.distance < 0.72*match2.distance:   #match1 better than match2 by 25% or less distance
        
        matches_mask[i] = [1,0]
        
```

```python
draw_params = dict(matchColor = (0,255,0), singlePointColor = (0,0,255), matchesMask = matches_mask, flags = 0)
```

```python
flann_matches = cv2.drawMatchesKnn(crunch,kp1,cereals,kp2,matches,None,**draw_params)
```

```python
display(flann_matches)
```
