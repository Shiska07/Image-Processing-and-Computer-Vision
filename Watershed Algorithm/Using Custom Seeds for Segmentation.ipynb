```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
view = cv2.imread('landscape.jpg')
view = cv2.resize(view,(960,600))
show_view = cv2.cvtColor(view,cv2.COLOR_BGR2RGB)
view_copy = np.copy(view)
```

```python
plt.figure(figsize = (10,10))
plt.imshow(show_view)
```

```python
# create empty space for the results to be drawn on
marker_image = np.zeros(view.shape[:2], dtype = np.int32)
```

```python
segments = np.zeros(view.shape,dtype = np.uint8)
```

```python
# create a function that will let you choose different colord from cm
def create_rgb(i):
    return tuple(np.array(cm.tab10(i)[:3])*255)
```

```python
colors = []
for i in range(10):
    colors.append(create_rgb(i))
```

```python
# create global variables
n_markers = 10           #0-9
current_marker = 1       #current color
marks_updated = False    #color changed
```

```python
# create callback function
def mouse_callback(event,x,y,flags,params):
    global current_marker, marks_updated
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        # marker passed to watershed algorithm
        cv2.circle(marker_image,(x,y),10,(current_marker),-1)
        
        # what user sees on the view
        cv2.circle(view_copy,(x,y),10,colors[current_marker],-1)
        marks_updated = True
        
```

```python
cv2.namedWindow('View')
cv2.setMouseCallback('View', mouse_callback)

while True:
    
    cv2.imshow('Watershed Segments',segments)
    cv2.imshow('View',view_copy)
    
    k = cv2.waitKey(1)
    
    if k == ord('q'):
        break
    
    # reset image
    elif k == ord('c'):
        view_copy = view.copy()
        marker_image = np.zeros(view.shape[:2], dtype = np.int32)
        segments = np.zeros(view.shape, dtype = np.uint8)
        
    # update color choice
    elif k > 0 and chr(k).isdigit():
        current_marker = int(chr(k))
        
    if marks_updated:
        marker_image_copy = marker_image.copy()
        cv2.watershed(view,marker_image_copy)
        
        segments = np.zeros(view.shape, dtype = np.uint8)
        
        for color_ind in range(n_markers):
            segments[marker_image_copy == (color_ind)] = colors[color_ind]
            
        marks_updated = False
        
cv2.destroyAllWindows()
```
