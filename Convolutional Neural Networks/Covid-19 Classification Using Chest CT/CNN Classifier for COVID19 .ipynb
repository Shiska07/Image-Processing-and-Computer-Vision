```python
# we need to make sure all images are the same size
from keras.preprocessing.image import ImageDataGenerator
```

```python
# this object creates more training images to make the network robust and invariant to some changes
image_gen = ImageDataGenerator(rotation_range = 15, 
                              width_shift_range = 0.05, 
                              height_shift_range = 0.15,
                              rescale = 1/255,
                              shear_range = 0.2,
                              zoom_range = 0.1,
                              horizontal_flip = False,
                              fill_mode = 'nearest')
```

```python
plt.imshow(image_gen.random_transform(cov4))
```

```python
from keras.models import Sequential
from keras.layers import Activation, Dropout, Flatten, Conv2D, MaxPool2D, Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# we will change all images to this size
input_shape = (1000,1000,3)

# kernel size
k_size = (3,3)
```

```python
# creating the model

model = Sequential()

model.add(Conv2D(filters = 16, kernel_size = k_size, input_shape = input_shape, activation = 'relu'))
model.add(MaxPool2D(pool_size = (2,2)))

model.add(Conv2D(filters = 32, kernel_size = k_size, input_shape = input_shape, activation = 'relu'))
model.add(MaxPool2D(pool_size = (2,2)))

model.add(Conv2D(filters = 32, kernel_size = k_size, input_shape = input_shape, activation = 'relu'))
model.add(MaxPool2D(pool_size = (2,2)))

model.add(Flatten())

model.add(Dense(64, activation = 'relu'))
model.add(Dropout(0.35))

model.add(Dense(32, activation = 'relu'))
model.add(Dropout(0.35))

model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# early stopping avoids data over fitting
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', patience = 4)
```

```python
model.summary()
```

```python
batch_size = 16

train_image_gen = image_gen.flow_from_directory('COVID19_DATA/train',
                                                target_size = input_shape[:2],
                                                batch_size = batch_size,
                                                class_mode = 'binary')
```

```python
test_image_gen = image_gen.flow_from_directory('COVID19_DATA/test',
                                                target_size = input_shape[:2],
                                                batch_size = batch_size,
                                                class_mode = 'binary')
```

```python
train_image_gen.class_indices
```

```python
results = model.fit_generator(train_image_gen,
                              epochs = 50,
                              validation_data = test_image_gen,
                              callbacks = [early_stop])
```

```python
model.save('COVID_Classifier.h5')
```

```python
# let's plot the validation_data loss
plt.plot(results.history['val_loss'])
```
