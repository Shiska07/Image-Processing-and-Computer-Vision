```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cat_tag = cv2.imread('cat_tag.png')
dog_tag = cv2.imread('dog_tag.png')
```

```python
cat_gray = cv2.cvtColor(cat_tag.copy(),cv2.COLOR_BGR2GRAY)
dog_gray = cv2.cvtColor(dog_tag.copy(),cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cat_gray, cmap='gray')
```

```python
res, thresh1 = cv2.threshold(cat_gray,130,255,cv2.THRESH_BINARY_INV)
res, thresh2 = cv2.threshold(dog_gray,130,255,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thresh1,cmap = 'gray')
```

```python
mask = cv2.bitwise_and(cat_tag,cat_tag,thresh1)
```

```python
plt.imshow(mask)
```
